# Median-normalization of the proteome raw numbers ([DESeq2 method](https://scienceparkstudygroup.github.io/research-data-management-lesson/median_of_ratios_manual_normalization/index.html))

```python
Project = '/Users/daehwa/Library/CloudStorage/OneDrive-Personal/Junlab/Projects/Adipocyte'
from IPython.display import display
import pandas as pd
import numpy as np
```

```python
MitoCarta3_file = '/Users/daehwa/Library/CloudStorage/OneDrive-Personal/Junlab/Data_Library/MITOCARTA3.0/Mouse.MitoCarta3.0.xls'
MitoCarta = pd.read_excel(MitoCarta3_file, 'A Mouse MitoCarta3.0')
MitoPW    = pd.read_excel(MitoCarta3_file, 'C MitoPathways')

name2EnsgID = MitoCarta[['Symbol','EnsemblGeneID']].set_index('Symbol')['EnsemblGeneID'].to_dict()
name2UniprotID = MitoCarta[['Symbol','UniProt']].set_index('Symbol')['UniProt'].to_dict()

MitoPW['PW level'] = MitoPW['MitoPathway Hierarchy'].str.split('>').str.len()
MitoPW['EnsemblGeneIDs'] = ''
MitoPW['UniProtIDs'] = ''
for i in range(len(MitoPW)):

    # Collecting Ensembl gene IDs
    for gene in MitoPW.iloc[i,:]['Genes'].split(', '):
        if type(name2EnsgID[gene]) == float : continue
        MitoPW.iloc[i,4] += name2EnsgID[gene].replace('|',',') + ','
    MitoPW.iloc[i,4] = MitoPW.iloc[i,4][:-1]

    # Collecting UniProt IDs
    for gene in MitoPW.iloc[i,:]['Genes'].split(', '):
        if type(name2UniprotID[gene]) == float : continue
        MitoPW.iloc[i,5] += name2UniprotID[gene].replace('|',',') + ','
    MitoPW.iloc[i,5] = MitoPW.iloc[i,5][:-1]

MitoPW = MitoPW.set_index('MitoPathway',drop=True)

display(MitoPW)


Ensembl_Genesets = {
    # 'OXPHOS complex' : pd.DataFrame({'gene_id' : MitoPW.loc['OXPHOS subunits','EnsemblGeneIDs'].split(',')}),
    # 'TCA cycle' : pd.DataFrame({'gene_id' : MitoPW.loc['TCA cycle','EnsemblGeneIDs'].split(',')}),
    # 'Lipid metabolism' : pd.DataFrame({'gene_id' : MitoPW.loc['Lipid metabolism','EnsemblGeneIDs'].split(',')}),
    # 'Fatty acid oxidation' : pd.DataFrame({'gene_id' : MitoPW.loc['Fatty acid oxidation','EnsemblGeneIDs'].split(',')}),
    # 'Mitochondrial ribosome' : pd.DataFrame({'gene_id' : MitoPW.loc['Mitochondrial ribosome','EnsemblGeneIDs'].split(',')}),
    'All MitoCarta' : MitoCarta[['EnsemblGeneID']].drop_duplicates().rename(columns={'EnsemblGeneID':'gene_id'})
}
Uniprot_Genesets = {
    # 'OXPHOS complex' : pd.DataFrame({'uniprot_id' : MitoPW.loc['OXPHOS subunits','UniProtIDs'].split(',')}),
    # 'TCA cycle' : pd.DataFrame({'uniprot_id' : MitoPW.loc['TCA cycle','UniProtIDs'].split(',')}),
    # 'Lipid metabolism' : pd.DataFrame({'uniprot_id' : MitoPW.loc['Lipid metabolism','UniProtIDs'].split(',')}),
    # 'Fatty acid oxidation' : pd.DataFrame({'uniprot_id' : MitoPW.loc['Fatty acid oxidation','UniProtIDs'].split(',')}),
    # 'Mitochondrial ribosome' : pd.DataFrame({'uniprot_id' : MitoPW.loc['Mitochondrial ribosome','UniProtIDs'].split(',')}),
    'All MitoCarta' : MitoCarta[['UniProt']].drop_duplicates().rename(columns={'UniProt':'uniprot_id'})
}
```

```python
PT_count = pd.read_excel(f'{Project}/Results/Other_papers/Martinez_etal.2023.Cell_Metabolism/Tables/Proteome/KBSI_DIA_NN_Dataprocessing(20240415)_20200731_LUM1_DIA_AfterNor_AfterANOVA_BAT_Beige_WAT.xlsx', sheet_name='2. Before Normal2')
PT_count = PT_count.drop(index=0, columns=['Unnamed: 0','Protein.Group','First.Gene','First.Protein.Description'])
PT_count = PT_count[~PT_count['First.Protein. (ProteinGroups의 가장 앞의 단백질)'].str.startswith('smORF')]

# Beige <-> White 순서 바뀐 것으로 보임
PT_count = PT_count.rename(columns={'First.Protein. (ProteinGroups의 가장 앞의 단백질)' : 'uniprot_id',
                                    'BAT_1856' : 'Brown_a',
                                    'BAT_1858' : 'Brown_b',
                                    'BAT_1859' : 'Brown_c',
                                    'BAT_1860' : 'Brown_d',
                                    'BAT_1861' : 'Brown_e',
                                    'Beige_1869' : 'White_a',
                                    'Beige_1870' : 'White_b',
                                    'Beige_1871' : 'White_c',
                                    'Beige_1872' : 'White_d',
                                    'Beige_1873' : 'White_e',
                                    'WAT_1880' : 'Beige_a',
                                    'WAT_1881' : 'Beige_b',
                                    'WAT_1882' : 'Beige_c',
                                    'WAT_1883' : 'Beige_d',
                                    'WAT_1885' : 'Beige_e',
                                    })

display(PT_count)
```

```python
PT_count_tmp = PT_count.copy()
PT_count_tmp.iloc[:,1:] = np.log(PT_count_tmp.iloc[:,1:])
PT_count_tmp = PT_count_tmp.replace(-np.inf, np.nan).dropna()
PT_count_tmp['pseudo_reference'] = PT_count_tmp.iloc[:,1:].mean(axis=1)
for SP in PT_count_tmp.iloc[:,1:]:
    PT_count_tmp[SP] = PT_count_tmp[SP] - PT_count_tmp['pseudo_reference']
PT_count_tmp = PT_count_tmp.drop(columns='pseudo_reference')
PT_scaling_factor = [np.exp(i) for i in PT_count_tmp.iloc[:,1:].median().tolist()]

display(PT_scaling_factor)
```

```python
PT_Ncount = PT_count.copy()
for SP, SF in zip(PT_Ncount.iloc[:,1:], PT_scaling_factor):
    PT_Ncount[SP] = PT_Ncount[SP] / SF

display(PT_Ncount)
```

```python
PT_log2Exp_rel = PT_Ncount.copy()
PT_log2Exp_rel.iloc[:,1:] = np.log2(PT_log2Exp_rel.iloc[:,1:])
for SP in PT_log2Exp_rel.iloc[:,1:]:
    PT_log2Exp_rel[SP] = PT_log2Exp_rel[SP] - PT_log2Exp_rel[SP].median()

display(PT_log2Exp_rel)
```

```python
# FC
Conds = [['Beige','Brown'], ['Beige','White'], ['Brown','White']]
reps = ['a','b','c','d','e']

PT_FC = pd.DataFrame()
PT_FC['uniprot_id'] = PT_Ncount['uniprot_id']
for Cond in Conds:
    for rep in reps:
        PT_FC[f'{Cond[0]}vs{Cond[1]}_'+rep] = PT_Ncount[Cond[0]+'_'+rep] / PT_Ncount[Cond[1]+'_'+rep]

# logFC
PT_logFC = PT_FC.copy()
PT_logFC.iloc[:,1:] = np.log2(PT_FC.iloc[:,1:])

# avg logFC
Conds = ['BeigevsBrown','BeigevsWhite','BrownvsWhite']
reps = ['a','b','c','d','e']

PT_avg_logFC = pd.DataFrame()
PT_avg_logFC['uniprot_id'] = PT_logFC['uniprot_id']
for Cond in Conds:
    PT_avg_logFC[Cond] = PT_logFC[[Cond+'_'+rep for rep in reps]].mean(axis=1)

display(PT_logFC, PT_avg_logFC)
```

```python
PT_count_mito = pd.merge(Uniprot_Genesets['All MitoCarta'], PT_count, on='uniprot_id')

display(PT_count_mito)
```

```python
PT_count_tmp = PT_count_mito.copy()
PT_count_tmp.iloc[:,1:] = np.log(PT_count_tmp.iloc[:,1:])
PT_count_tmp = PT_count_tmp.replace(-np.inf, np.nan).dropna()
PT_count_tmp['pseudo_reference'] = PT_count_tmp.iloc[:,1:].mean(axis=1)
for SP in PT_count_tmp.iloc[:,1:]:
    PT_count_tmp[SP] = PT_count_tmp[SP] - PT_count_tmp['pseudo_reference']
PT_count_tmp = PT_count_tmp.drop(columns='pseudo_reference')
PT_scaling_factor = [np.exp(i) for i in PT_count_tmp.iloc[:,1:].median().tolist()]

display(PT_scaling_factor)
```

```python
PT_Ncount_mito = PT_count_mito.copy()
for SP, SF in zip(PT_Ncount_mito.iloc[:,1:], PT_scaling_factor):
    PT_Ncount_mito[SP] = PT_Ncount_mito[SP] / SF

display(PT_Ncount_mito)
```

```python
PT_log2Exp_rel_mito = PT_Ncount_mito.copy()
PT_log2Exp_rel_mito.iloc[:,1:] = np.log2(PT_log2Exp_rel_mito.iloc[:,1:])
for SP in PT_log2Exp_rel_mito.iloc[:,1:]:
    PT_log2Exp_rel_mito[SP] = PT_log2Exp_rel_mito[SP] - PT_log2Exp_rel_mito[SP].median()

display(PT_log2Exp_rel_mito)
```

```python
# FC
Conds = [['Beige','Brown'], ['Beige','White'], ['Brown','White']]
reps = ['a','b','c','d','e']

PT_FC_mito = pd.DataFrame()
PT_FC_mito['uniprot_id'] = PT_Ncount_mito['uniprot_id']
for Cond in Conds:
    for rep in reps:
        PT_FC_mito[f'{Cond[0]}vs{Cond[1]}_'+rep] = PT_Ncount_mito[Cond[0]+'_'+rep] / PT_Ncount_mito[Cond[1]+'_'+rep]

# logFC
PT_logFC_mito = PT_FC_mito.copy()
PT_logFC_mito.iloc[:,1:] = np.log2(PT_FC_mito.iloc[:,1:])

# avg logFC
Conds = ['BeigevsBrown','BeigevsWhite','BrownvsWhite']
reps = ['a','b','c','d','e']

PT_avg_logFC_mito = pd.DataFrame()
PT_avg_logFC_mito['uniprot_id'] = PT_logFC_mito['uniprot_id']
for Cond in Conds:
    PT_avg_logFC_mito[Cond] = PT_logFC_mito[[Cond+'_'+rep for rep in reps]].mean(axis=1)

display(PT_logFC_mito, PT_avg_logFC_mito)
```
