```python
Project = '/Users/daehwa/Library/CloudStorage/OneDrive-Personal/Junlab/Projects/Adipocyte'

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns

plt.rcParams.update(plt.rcParamsDefault)

mpl.rcParams['font.sans-serif'] = "Arial"
mpl.rcParams['font.family'] = "sans-serif"

plt.rcParams['ps.useafm']=True
plt.rcParams['pdf.fonttype']=42

plt.rc('font', size=7)          # controls default text sizes
plt.rc('axes', titlesize=8)     # fontsize of the axes title
plt.rc('axes', labelsize=8)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=6)    # fontsize of the tick labels
plt.rc('ytick', labelsize=6)    # fontsize of the tick labels
plt.rc('legend', fontsize=6)    # legend fontsize
suptitlesize = 8

plt.rc('axes', axisbelow=True) #push axes to back
mpl.rcParams['axes.linewidth'] = 0.5
```

```python
Metadata = pd.read_csv(f'{Project}/Results/Other_papers/Emont_etal.2022.Nature/scRNAseq/scExp_data_slct1.tsv.gz', sep='\t', compression='gzip')
display(Metadata)
display(set(Metadata['cell_type']))

sub_Metadata = pd.merge(Metadata, pd.DataFrame({'cell_type':['ASPC','Adipocyte']}), on='cell_type')[::-1]
display(sub_Metadata)
```

```python
sF = 1.15 # Size factor
wgap = 0.35
hgap = 0.3
nrow, ncol = 2, 2

width_ratios = [1, 0.5]

fs_w = sum(np.array(width_ratios)*sF) + (ncol-1)*wgap*sF
fs_h = nrow*sF + (nrow-1)*hgap*sF

#
fig = plt.figure(figsize=( fs_w, fs_h ), dpi=300)
grid = plt.GridSpec(nrow, ncol,
                    width_ratios=width_ratios,
                    left=0, right=1, bottom=0, top=1,
                    wspace=wgap, hspace=hgap)

# Gls UMAP
ax = fig.add_subplot(grid[0,0])

sct = ax.scatter(Metadata['UMAP_1'], Metadata['UMAP_2'],
                c=Metadata['Gls'], cmap="Reds",
                s=0.05, alpha=1, lw=0, rasterized=True,
                # vmin=0, vmax=4
                )

ax.set_title('Gls', fontsize=7)
ax.set_xlabel('UMAP 1', labelpad=0, fontsize=6)
ax.set_ylabel('UMAP 2', labelpad=0, fontsize=6)
ax.set_xticks([])
ax.set_yticks([])

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
# ax.spines['bottom'].set_visible(False)
# ax.spines['left'].set_visible(False)

cbar = plt.colorbar(sct, ax=ax, cax = ax.inset_axes([0.93, 0.03, 0.05, 0.35]), orientation='vertical')
# cbar.ax.set_ylabel('Normalized\nExpression', labelpad=1, fontsize=6)
cbar.ax.yaxis.tick_left()
cbar.ax.yaxis.set_label_position('left')
cbar.ax.set_yticks([0,3])
cbar.ax.tick_params(width=0.5, length=2.5, pad=1)


# Gls violin
ax = fig.add_subplot(grid[0,1])

sns.violinplot(data=sub_Metadata,
               x='cell_type',
               y='Gls', 
               palette=['#FFCC99','#E69F00'],
               scale='width',
               inner=None,
               linewidth=0.5)

# ax.set_title('Gls', fontsize=7)
ax.set_xlabel('', labelpad=0, fontsize=6)
ax.set_ylabel('Normalized Expression', labelpad=0, fontsize=6)
ax.set_xticklabels(['ASPC','Adipocyte'], rotation=-30, ha='left', va='top')
# ax.set_yticks([])
ax.tick_params(width=0.5, length=3, pad=2)

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
# ax.spines['bottom'].set_visible(False)
# ax.spines['left'].set_visible(False)



# Glul UMAP
ax = fig.add_subplot(grid[1,0])

sct = ax.scatter(Metadata['UMAP_1'], Metadata['UMAP_2'],
                c=Metadata['Glul'], cmap="Reds",
                s=0.05, alpha=1, lw=0, rasterized=True,
                # vmin=0, vmax=4
                )

ax.set_title('Glul', fontsize=7, y=0.97)
ax.set_xlabel('UMAP 1', labelpad=0, fontsize=6)
ax.set_ylabel('UMAP 2', labelpad=0, fontsize=6)
ax.set_xticks([])
ax.set_yticks([])

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
# ax.spines['bottom'].set_visible(False)
# ax.spines['left'].set_visible(False)

cbar = plt.colorbar(sct, ax=ax, cax = ax.inset_axes([0.93, 0.03, 0.05, 0.35]), orientation='vertical')
# cbar.ax.set_ylabel('Normalized\nExpression', labelpad=1, fontsize=6)
cbar.ax.yaxis.tick_left()
cbar.ax.yaxis.set_label_position('left')
cbar.ax.set_yticks([0,4])
cbar.ax.tick_params(width=0.5, length=2.5, pad=1)


# Glul violin
ax = fig.add_subplot(grid[1,1])

sns.violinplot(data=sub_Metadata,
               x='cell_type',
               y='Glul', 
               palette=['#FFCC99','#E69F00'],
               scale='width',
               inner=None,
               linewidth=0.5)

# ax.set_title('Glul', fontsize=7)
ax.set_xlabel('', labelpad=0, fontsize=6)
ax.set_ylabel('Normalized Expression', labelpad=0, fontsize=6)
ax.set_xticklabels(['ASPC','Adipocyte'], rotation=-30, ha='left', va='top')
# ax.set_yticks([])
ax.tick_params(width=0.5, length=3, pad=2)

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
# ax.spines['bottom'].set_visible(False)
# ax.spines['left'].set_visible(False)


plt.savefig('Gls_Glul_UMAP_violin.pdf', bbox_inches='tight')
plt.show()    
```
