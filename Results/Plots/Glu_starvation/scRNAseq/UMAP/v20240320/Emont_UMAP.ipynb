```python
Metadata = pd.read_csv(f'{Project}/Results/Other_papers/Emont_etal.2022.Nature/scRNAseq/Emont_UMAP_metadata1.tsv', sep='\t')
display(Metadata)
display(set(Metadata['cell_type']))
```

```python
custom_order = ['Endothelial', 'Epithelial', 'Immune cells', 'Mesothelial', 'Smooth muscle'] + ['ASPC', 'Adipocyte']

color_order = {'ASPC'          : '#FFCC99', 
               'Adipocyte'     : '#E69F00', 
               'Endothelial'   : '#AAAAAA', 
               'Epithelial'    : '#AAAAAA', 
               'Immune cells'  : '#AAAAAA', 
               'Mesothelial'   : '#AAAAAA', 
               'Smooth muscle' : '#AAAAAA'}
```

```python
fig = plt.figure(figsize=(1.8, 1.8), dpi=150)
ax = plt.axes([0,0,1,1])

for order in custom_order:
    tmp_data = Metadata[Metadata['cell_type']==order]
    ax.scatter(tmp_data['UMAP_1'], tmp_data['UMAP_2'], label=order,
                s=0.1, alpha=1, lw=0, color=color_order[order], rasterized=True)

ax.set_title('')
ax.set_xlabel('UMAP 1', labelpad=2)
ax.set_ylabel('UMAP 2', labelpad=2)
ax.set_xticks([])
ax.set_yticks([])

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
# ax.spines['bottom'].set_visible(False)
# ax.spines['left'].set_visible(False)

# lgnd = plt.legend(bbox_to_anchor=(1,1), loc="upper left")
# for i in range(len(custom_order[filt])): lgnd.legendHandles[i]._sizes = [70]

# plt.savefig('Emont_UMAP_type1.pdf', bbox_inches='tight')
plt.show()    
```
